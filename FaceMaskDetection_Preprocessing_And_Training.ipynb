```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
```

```python
def build_model(hp):
  model = keras.Sequential([
    keras.layers.Conv2D(
        filters=hp.Int('conv_1_filter', min_value=32, max_value=128, step=16),
        # filters=32,
        kernel_size=hp.Choice('conv_1_kernel', values = [3,5]),
        activation='relu',
        input_shape=(96,96,3)
    ),
    layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(
        filters=hp.Int('conv_2_filter', min_value=16, max_value=64, step=8),
        # filters=32,
        kernel_size=hp.Choice('conv_2_kernel', values = [5,7]),
        activation='relu'
    ),
    layers.MaxPooling2D((2, 2)),
    keras.layers.Flatten(),
    keras.layers.Dense(
        # units=hp.Int('dense_1_units', min_value=32, max_value=256, step=32),
        units=64,
        activation='relu'
    ),
    keras.layers.Dense(
        # units=hp.Int('dense_1_units', min_value=32, max_value=256, step=32),
        units=64,
        activation='relu'
    ),
    keras.layers.Dense(2, activation='softmax')
  ])
  
  model.compile(optimizer=keras.optimizers.Adam(hp.Choice('learning_rate', values=[1e-3, 1e-4])),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
  
  return model
```

```python
from keras_tuner import RandomSearch
from keras_tuner.engine.hyperparameters import HyperParameters
```

```python
tuner_search=RandomSearch(build_model,
                          objective='val_accuracy',
                          max_trials=5,directory='output',project_name="FaceMaskDetection")
```

```python
tuner_search.search(train_images,train_labels,epochs=3,validation_data=(valid_images, valid_labels))
```

```python
model=tuner_search.get_best_models(num_models=1)[0]
```

```python
model.summary()
```

```python
from keras import callbacks
earlystopping = callbacks.EarlyStopping(monitor ="val_loss", 
                                        mode ="min", patience = 5, 
                                        restore_best_weights = True)
```

```python
history = model.fit(train_images, train_labels, 
                    epochs = 25, validation_data=(valid_images, valid_labels), 
                    callbacks =[earlystopping])
```

```python
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.01, 0.1])
plt.legend(loc='lower right')
```

```python
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print(test_loss, ' ', test_acc)
```
